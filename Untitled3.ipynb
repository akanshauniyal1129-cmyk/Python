```python
# List

customer=["Sanjana","Mohit","Disha","Rohit","Vanita","Kartik","Nidhi","Sahil","Sunita","Dhruv"]
state=["UK","Punjab","Delhi","Mumbai","UK","UP","Goa","Kerela","Punjab","Assam"]
product=["Headphone","Watch","Phone","Laptop","Shrit","Headphone","Purse","Rings","Watch","Phone"]
amount=[12000,13000,14000,15000,16000,17000,18000,19000,10000,11000]
gender=["Female","Male","Female","Male","Female","Male","Female","Male","Female","Male"]
profit=[5,10,15,20,25,30,35,40,45,50]
print("Customer:-",customer)
print("State:-",state)
print("Product",product)
print("Amount:-",amount)
print("Gender:-",gender)
print("Profit:-",profit)

# Total Amount
total=0
for i in amount:
  total+=i
print("Total:-",total)

# Total Profit
total_profit=0
for i in profit:
  total_profit+=i
print("Total_Profit:-",total_profit)

# State Wise Sale
unique_state=[]
state_wise_sale=[]
for i in state:
  if i not in unique_state:
    unique_state.append(i)
    state_wise_sale.append(0)
for i in range(len(state)):
    index=unique_state.index(state[i])
    state_wise_sale[index]+=amount[i]
print("Unique_State:-",unique_state)
print("State_Wise_Product",state_wise_sale)

# Product Wise Sale
unique_product=[]
product_wise_sale=[]
for i in product:
  if i not in unique_product:
    unique_product.append(i)
    product_wise_sale.append(0)
for i in range(len(product)):
  index=unique_product.index(product[i])
  product_wise_sale[index]+=amount[i]
print("Unique_Product:-",unique_product)
print("Product_Wise_Sale:-",product_wise_sale)

# Gender Wise Sale
unique_gender=[]
gender_wise_sale=[]
for i in gender:
  if i not in unique_gender:
    unique_gender.append(i)
    gender_wise_sale.append(0)
for i in range(len(gender)):
  index=unique_gender.index(gender[i])
  gender_wise_sale[index]+=amount[i]
print("Unique_Gender:-",unique_gender)
print("Gender_Wise_Sale:-",gender_wise_sale)

# State Wise Profit
unique_state=[]
state_wise_profit=[]
for i in state:
  if i not in unique_state:
    unique_state.append(i)
    state_wise_profit.append(0)
for i in range(len(state)):
    index=unique_state.index(state[i])
    state_wise_profit[index]+=profit[i]
print("Unique_State:-",unique_state)
print("State_Wise_Profit:-",state_wise_profit)

# Product Wise Profit
unique_product=[]
product_wise_profit=[]
for i in product:
  if i not in unique_product:
    unique_product.append(i)
    product_wise_profit.append(0)
for i in range(len(product)):
  index=unique_product.index(product[i])
  product_wise_profit[index]+=profit[i]
print("Unique_Product:-",unique_product)
print("Product_wise_profit:-",product_wise_profit)

# Gender Wise Profit
unique_gender=[]
gender_wise_profit=[]
for i in gender:
  if i not in unique_gender:
    unique_gender.append(i)
    gender_wise_profit.append(0)
for i in range(len(gender)):
  index=unique_gender.index(gender[i])
  gender_wise_profit[index]+=profit[i]
print("Unique_gender:-",unique_gender)
print("Gender_wise_Profit:-",gender_wise_profit)

# Graph
import matplotlib.pyplot as plt
colors=["Red","Blue","Green"]
plt.bar(unique_state,state_wise_sale,color=colors)
plt.xlabel("unique_state")
plt.ylabel("state_wise_sale")
plt.title("State Wise Sales")
plt.show()
plt.bar(unique_product,product_wise_sale)
plt.xlabel("unique_product")
plt.ylabel("product_wise_sale")
plt.title("Product Wise Sale")
plt.show()
plt.bar(unique_gender,gender_wise_sale)
plt.xlabel("unique_gender")
plt.ylabel("gender_wise_sale")
plt.title("Gender Wise Sale")
plt.show()
plt.bar(unique_state,state_wise_profit)
plt.xlabel("unique_state")
plt.ylabel("state_wise_profit")
plt.title("State Wise Profit")
plt.show()
plt.bar(unique_product,product_wise_profit)
plt.xlabel("unique_product")
plt.ylabel("product_wise_profit")
plt.title("Product Wise Profit")
plt.show()
plt.bar(unique_gender,gender_wise_profit)
plt.xlabel("unique_gender")
plt.ylabel("gender_wise_profit")
plt.title("Gender Wise Profit")
plt.show()
```

```python
import csv
Customer=[]
Gender=[]
State=[]
Product=[]
Amount=[]
with open("sales.csv",'r') as file:
  reader=csv.reader(file)
  header=next(reader)
  for i in reader:
    Customer.append(i[0])
    Gender.append(i[1])
    State.append(i[2])
    Product.append(i[3])
    Amount.append(i[4])

  # Gender Wise

  unique_gender=[]
  gender_wise=[]
  for i in Gender:
    if i not in unique_gender:
      unique_gender.append(i)
      gender_wise.append(0)
  for i in range(len(Gender)):
    index=unique_gender.index(Gender[i])
    gender_wise[index]+=int(Amount[i])
  print("Unique_Wise:-",unique_gender)
  print("Gender_Wise:-",gender_wise)

  # State Wise

  unique_state=[]
  state_wise=[]
  for i in State:
    if i not in unique_state:
      unique_state.append(i)
      state_wise.append(0)
  for i in range(len(State)):
    index=unique_state.index(State[i])
    state_wise[index]+=int(Amount[i])
  print("Unique_State:-",unique_state)
  print("State_Wise:-",state_wise)

  # Product Wise

  unique_product=[]
  product_wise=[]
  for i in Product:
    if i not in unique_product:
      unique_product.append(i)
      product_wise.append(0)
  for i in range(len(Product)):
    index=unique_product.index(Product[i])
    product_wise[index]+=int(Amount[i])
  print("Unique_Product:-",unique_product)
  print("Product_Wise:-",product_wise)

import matplotlib.pyplot as plt
plt.bar(unique_gender,gender_wise)
plt.xlabel("Unique_Gender")
plt.ylabel("Gender_Wise")
plt.title("Gender Wise")
plt.show()

plt.bar(unique_state,state_wise)
plt.xlabel("Unique_State")
plt.ylabel("State_Wise")
plt.title("State Wise")
plt.show()

plt.bar(unique_product,product_wise)
plt.xlabel("Unique_Product")
plt.ylabel("Product_Wise")
plt.title("Product Wise")
plt.show()
```

```python
Country=["India","USA","Japan","China","India","Germeny","USA","Landon","Japan","Genmeny"]
Cases=[48169,21832,23344,34518,24402,48169,21832,23344,34518,24402]
Deaths=[470,502,404,669,321,532,470,669,532,502]
Recovered=[41211,10627,9684,27701,11301,9684,11301,10627,41211,10627]
Active_Cases=[3456,2341,56453,2341,67453,56353,3456,56453,67453,54643]
Vaccinations=[546,435,356,768,876,986,546,435,356,768]

# Country Wise Cases

unique_country=[]
country_wise_cases=[]
for i in Country:
  if i not in unique_country:
    unique_country.append(i)
    country_wise_cases.append(0)
for i in range(len(Country)):
  index=unique_country.index(Country[i])
  country_wise_cases[index]+=Cases[i]
print("Unique_Country:-",unique_country)
print("Country_Wise_Cases:-",country_wise_cases)

# Country Wise Death

unique_country=[]
country_wise_death=[]
for i in Country:
  if i not in unique_country:
    unique_country.append(i)
    country_wise_death.append(0)
for i in range(len(Country)):
  index=unique_country.index(Country[i])
  country_wise_death[index]+=Deaths[i]
print("Unique_Country:-",unique_country)
print("Country_Wise_Death:-",country_wise_death)

# Country Wise Recovered

unique_country=[]
country_wise_recovery=[]
for i in Country:
  if i not in unique_country:
    unique_country.append(i)
    country_wise_recovery.append(0)
for i in range(len(Country)):
  index=unique_country.index(Country[i])
  country_wise_recovery[index]+=Recovered[i]
print("Unique_Country:-",unique_country)
print("Country_Wise_Recovery:-",country_wise_recovery)

# Country Wise Active Cases

unique_country=[]
country_wise_active_cases=[]
for i in Country:
  if i not in unique_country:
    unique_country.append(i)
    country_wise_active_cases.append(0)
for i in range(len(Country)):
  index=unique_country.index(Country[i])
  country_wise_active_cases[index]+=Recovered[i]
print("Unique_Country:-",unique_country)
print("Country_Wise_Active_Cases:-",country_wise_active_cases)

# Country Wise Vaccinations

unique_country=[]
country_wise_vaccination=[]
for i in Country:
  if i not in unique_country:
    unique_country.append(i)
    country_wise_vaccination.append(0)
for i in range(len(Country)):
  index=unique_country.index(Country[i])
  country_wise_vaccination[index]+=Recovered[i]
print("Unique_Country:-",unique_country)
print("Country_Wise_Vaccinations:-",country_wise_vaccination)

# Graph

import matplotlib.pyplot as plt
plt.bar(unique_country,country_wise_cases)
plt.xlabel("Unique_Country")
plt.ylabel("Country_Wise_Cases")
plt.title("Country Wise Cases")
plt.show()
```

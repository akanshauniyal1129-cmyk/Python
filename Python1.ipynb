```python
n=int(input("Enter number of products: "))
Name=[]
Products=[]
States=[]
Months=[]
Prices=[]
Orders=[]
Profits=[]
Sales=[]
for i in range(1,n+1):
  name=input(f"Enter Name {i}: ")
  Product=input(f"Enter Products {i}: ")
  State=input(f"Enter States {i}: ")
  Month=input(f"Enter Months {i}: ")
  Price=int(input(f"Enter Prices {i}: "))
  Order=int(input(f"Enter Orders {i}: "))
  Profit=int(input(f"Enter Profits {i}: "))
  Sale=int(input(f"Enter Sales {i}: "))
  Name.append(name)
  Products.append(Product)
  States.append(State)
  Months.append(Month)
  Prices.append(Price)
  Orders.append(Order)
  Profits.append(Profit)
  Sales.append(Sale)
max_Product=Products[0]
max_Sale=Sales[0]
max_state=States[0]
max_month=Months[0]
for i in range(1,n):
  if Sales[i] > max_Sale:
    max_Product=Products[i]
    max_Sale=Sales[i]
    max_state=States[i]
    max_month=Months[i]
print("Name:-",Name)
print("Product:-",Products)
print("State:-",States)
print("Month:-",Months)
print("Price:-",Prices)
print("Order:-",Orders)
print("Profit:-",Profits)
print("Sales:-",Sales)
print("Max_Product:-",max_Product)
print("Max_Sale:-",max_Sale)
print("Max_state:-",max_state)
print("Max_Month:-",max_month)
```

```python
from __future__ import with_statement
import csv
Customer=[]
Gender=[]
State=[]
Product=[]
Amount=[]
with open("sales.csv",'r') as file:
  reader=csv.reader(file)
  header=next(reader)
  for i in reader:
    Customer.append(i[0])
    Gender.append(i[1])
    State.append(i[2])
    Product.append(i[3])
    Amount.append(i[4])

  # Gender Wise

  unique_gender=[]
  gender_wise=[]
  for i in Gender:
    if i not in unique_gender:
      unique_gender.append(i)
      gender_wise.append(0)
  for i in range(len(Gender)):
    index=unique_gender.index(Gender[i])
    gender_wise[index]+=int(Amount[i])
  print("Unique_Wise:-",unique_gender)
  print("Gender_Wise:-",gender_wise)

  # State Wise

  unique_state=[]
  state_wise=[]
  for i in State:
    if i not in unique_state:
      unique_state.append(i)
      state_wise.append(0)
  for i in range(len(State)):
    index=unique_state.index(State[i])
    state_wise[index]+=int(Amount[i])
  print("Unique_State:-",unique_state)
  print("State_Wise:-",state_wise)

  # Product Wise

  unique_product=[]
  product_wise=[]
  for i in Product:
    if i not in unique_product:
      unique_product.append(i)
      product_wise.append(0)
  for i in range(len(Product)):
    index=unique_product.index(Product[i])
    product_wise[index]+=int(Amount[i])
  print("Unique_Product:-",unique_product)
  print("Product_Wise:-",product_wise)

import matplotlib.pyplot as plt
plt.bar(unique_gender,gender_wise)
plt.xlabel("Unique_Gender")
plt.ylabel("Gender_Wise")
plt.title("Gender Wise")
plt.show()

plt.bar(unique_state,state_wise)
plt.xlabel("Unique_State")
plt.ylabel("State_Wise")
plt.title("State Wise")
plt.show()

plt.bar(unique_product,product_wise)
plt.xlabel("Unique_Product")
plt.ylabel("Product_Wise")
plt.title("Product Wise")
plt.show()
```
